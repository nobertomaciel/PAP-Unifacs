```python
# PSEUDOCÓDIGO (while) -------------------

# var nota1, nota2: real;
# Início
# 	var contador: inteiro;
# 	contador = 1;
# 	Enquanto contador <= 10 Faça
# 		Ler nota1;
# 		Ler nota2;
# 		Escrever (nota1 + nota2)/2;
# 		contador = contador + 1;
# 	Fim Enquanto
# Fim

# CÓDIGO EM PYTHON  -------------------
nota1 = 0
nota2 = 0
contador = 1
while contador <= 5:
 print("iteração: ", contador)
 nota1 = int(input('Digite a primeira nota: '))
 nota2 = int(input('Digite a segunda nota: '))
 print('Média do aluno ',contador, ':', (nota1 + nota2)/2)
 contador = contador + 1
```

```python
# PSEUDOCÓDIGO (do while) -------------------

# var nota1, nota2: real;
# Início
# 	var contador: inteiro;
# 	contador = 1;
# 	Faça
# 		Ler nota1;
# 		Ler nota2;
# 		Escrever (nota1 + nota2)/2;
# 		contador = contador + 1;
# 	Enquanto contador <= 10 Faça
# Fim

# CÓDIGO EM PYTHON  -------------------
contador = 1
while True:
 print(contador)
 nota1 = int(input('Digite a primeira nota: '))
 nota2 = int(input('Digite a segunda nota: '))
 print('Média do aluno ',contador, ':', (nota1 + nota2)/2)
 contador = contador + 1
 if contador > 5:
   break
```

```python
# PSEUDOCÓDIGO (for) -------------------

# var nota1, nota2: real;
# Início
# 	var contador: inteiro;
# 	Para contador=0 até 5 Faça
# 		Ler nota1;
# 		Ler nota2;
# 		Escrever (nota1 + nota2)/2;
# 		contador = contador + 1;
# 	Fim Para
# Fim

# CÓDIGO EM PYTHON  -------------------
contador = [1,2,3,4,5]
for x in contador:
  print(x)
  nota1 = int(input('Digite a primeira nota: '))
  nota2 = int(input('Digite a segunda nota: '))
  print('Média do aluno ',x, ':', (nota1 + nota2)/2)
```
